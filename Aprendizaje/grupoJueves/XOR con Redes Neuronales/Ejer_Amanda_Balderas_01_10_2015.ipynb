```python
import pandas as pd
import numpy as np
```

Generamos valores para x1 y x2.

```python
datos = pd.DataFrame(np.random.randint(0, 2, 10000), columns=['x1'])
datos['x2'] = np.random.randint(0, 2, 10000)
datos.head()
```

Generamos los valores de y.

```python
Y = []
for i in range(len(datos)):
    if datos.iloc[i][0] == 0 and datos.iloc[i][1] == 0:
        Y.insert(i, 0)
    elif datos.iloc[i][0] == 0 and datos.iloc[i][1] == 1:
        Y.insert(i, 0)
    elif datos.iloc[i][0] == 1 and datos.iloc[i][1] == 0:
        Y.insert(i, 0)
    else:
        Y.insert(i, 1)
datos['y'] = Y
datos.head()
```

```python
from pybrain.tools.shortcuts import buildNetwork
from pybrain.datasets import SupervisedDataSet
from pybrain.supervised.trainers import BackpropTrainer
```

```python
net = buildNetwork(2, 2, 1)
```

```python
ds = SupervisedDataSet(2, 1)
```

```python
x = np.arange(0.0, 1.0+0.01, 0.01)
x
```

```python
Xs1 = datos[['x1', 'x2']]
Xs = []
for i in range(len(Xs1)):
    Xs.insert(i, tuple(Xs1.iloc[i]))
```

```python
Xs = np.array(Xs)
Ys = np.array(datos['y'])
```

```python
ds.setField('input', Xs.reshape(len(Xs),2))
```

```python
ds.setField('target', Ys.reshape(len(Ys),1))
```

```python
trainer = BackpropTrainer(net, ds)
```

```python
trainer.train()
```

```python
trainers = []
for i in range(5):
    trainers.insert(i, trainer.train())
```

```python
trainers
```

```python
for input, target in ds:
     print[net.activate(input), target]
```

```python
errores = trainer.trainUntilConvergence() 
errores
```
